```python
total_players = purchase_df.loc[:,['SN','Gender','Price','Age']]
total_players_count_l = list(total_players['SN'].unique())
total_players_count = len(total_players_count_l)


total_players_df = pd.DataFrame({'Total Players':[total_players_count]})
total_players_df.head()
```

```python
gender_l = total_players.copy()


    
# gender_total = len(gender_l)
# type(gender_l)
gender_l.head()
```

```python
columns = ['SN','Gender','Price']
male_count_df = gender_l.loc[gender_l['Gender']=='Male',columns]
female_count_df = gender_l.loc[gender_l['Gender']=='Female',columns]
other_count_df = gender_l.loc[gender_l['Gender']=='Other / Non-Disclosed',columns] 

#male_count_df
#print(female_count)
#print(other_count)
```

```python
unique_male_count = male_count_df['SN'].value_counts()
total_unique_male_count = len(unique_male_count)
total_unique_male_count

unique_female_count = female_count_df['SN'].value_counts()
total_unique_female_count = len(unique_female_count)
total_unique_female_count

unique_other_count = other_count_df['SN'].value_counts()
total_unique_other_count = len(unique_other_count)
total_unique_other_count

summary_gender_table_df = pd.DataFrame({'Gender':['Male','Female','Other / Non-Disclosed'],
                                    'Total Count':[total_unique_male_count,total_unique_female_count,total_unique_other_count],
                                    'Percentage of Players':[total_unique_male_count/total_players_count,
                                                            total_unique_female_count/total_players_count,
                                                            total_unique_other_count/total_players_count]})

summary_gender_table_df['Percentage of Players']= summary_gender_table_df['Percentage of Players'].map('{:,.2%}'.format)
summary_gender_table_df.set_index(['Gender'])
```

```python
male_purchase_count = len(male_count_df)
female_purchase_count = len(female_count_df)
other_purchase_count = len(other_count_df)

male_purchase_avg = male_count_df['Price'].mean()
female_purchase_avg = female_count_df['Price'].mean()
other_purchase_avg = other_count_df['Price'].mean()

tot_male_purchase = male_count_df['Price'].sum()
tot_female_purchase = female_count_df['Price'].sum()
tot_other_purchase = other_count_df['Price'].sum()

avg_tot_male_purchase = tot_male_purchase/total_unique_male_count
avg_tot_female_purchase = tot_female_purchase/total_unique_female_count
avg_tot_other_purchase = tot_other_purchase/total_unique_other_count

purchase_analysis_df = pd.DataFrame({'Gender':['Female','Male','Other / Non-Disclosed'],
                                    'Purchase Count':[female_purchase_count,male_purchase_count,other_purchase_count],
                                    'Average Purchase Price':[female_purchase_avg,male_purchase_avg,other_purchase_avg],
                                    'Total Purchase Value':[tot_female_purchase,tot_male_purchase,tot_other_purchase],
                                    'Avg Total Purchase per Person':[avg_tot_female_purchase,avg_tot_male_purchase,avg_tot_other_purchase]})

purchase_analysis_df['Average Purchase Price'] = purchase_analysis_df['Average Purchase Price'].map('${:,.2f}'.format)
purchase_analysis_df['Total Purchase Value'] = purchase_analysis_df['Total Purchase Value'].map('${:,.2f}'.format)
purchase_analysis_df['Avg Total Purchase per Person'] = purchase_analysis_df['Avg Total Purchase per Person'].map('${:,.2f}'.format)
purchase_analysis_df.set_index(['Gender'])
```

```python
bins = [0,9,14,19,24,29,34,39,100]
bin_labels = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']

unique_sn = total_players.drop_duplicates(subset="SN")
unique_sn['Age Ranges'] = pd.cut(unique_sn['Age'],bins,labels=bin_labels)
unique_sn.set_index('Age Ranges')

unique_sn = unique_sn.groupby(['Age Ranges'])

unique_count_age_group=unique_sn['Gender'].count()
#unique_count_age_group

percentage_l = [row for row in unique_count_age_group/total_players_count]
#percentage_l

age_demo_df = pd.DataFrame({'Total Count':unique_count_age_group,'Percentage of Players':percentage_l})
age_demo_df['Percentage of Players']=age_demo_df['Percentage of Players'].map('{:.2%}'.format)
age_demo_df
```
